```python
from numpy import *

A = zeros([4,4])
rows, shape = A.shape

b = zeros([4,1])

row_sum = 0
#Fill the 4*4 matrix  
for i in range(1,5):
    for j in range(1,5):
        A[i-1,j-1] = 1/(100*i + 0.01) + j;
        row_sum += A[i-1,j-1];
    b[i-1, 0] = row_sum;

#Compute SVD
U, sigma, Vt = linalg.svd(A);

print("A: \n", A,"\n");
print("b: \n", b, "\n");

#find the unknown of Ax = b
def getUnknownOfA(): 
    Sigma_pinv = zeros([4,4]);
    Sigma_pinv = diag(1/sigma)
    
    x = Vt.T.dot(Sigma_pinv.dot(U.T.dot(b)));
    return x;
    
x1 = getUnknownOfA();
print("X1: Ax = b\n", x1)

#set zero to the smallest singular value
sigma[1] = 0;

#move the zero singular value to last column and change the position of Vt
sigma[[3,1]] = sigma[[1,3]];
Vt[[3,1]] = Vt[[1,3]];

#Make sigma as diagonal
Sigma = zeros([4,1]);
Sigma = diag(sigma)

#B Matrix has smallest SV as zero
B = U.dot(Sigma.dot(Vt));
print("Matrix B has set the smallest singular value to zero");
print("Matrix B: \n", B, "\n");

#Compute the rank of A and B matrices
print("Rank A :", linalg.matrix_rank(A));
print("Rank B :", linalg.matrix_rank(B));

#find the unknown of Bx = b
def getUnknownOfB(): 
    Sigma_pinv = zeros([4,4]);
    Sigma_pinv[:3,:3] = diag(1/sigma[:3])
    
    x = Vt.T.dot(Sigma_pinv.dot(U.T.dot(b)));
    return x;
    
x2 = getUnknownOfB();
print("x2: Bx = b \n", x2)
```
