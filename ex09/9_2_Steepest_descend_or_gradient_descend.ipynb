```python
import matplotlib.pyplot as plt
import numpy as np

#loss function 
function = lambda x : -3.325 * (x ** 2)

#Get 500 evenly spaced numbers between -3 and 3 (arbitratil chosen to ensure steep curve)
x = np.linspace(-3,3,500)

#Plot the curve
plt.plot(x, function(x))
plt.show()
```

```python

cur_x = 1 # The algorithm starts at x=3
rate = 0.005 # Learning rate
precision = 0.001 #This tells us when to stop the algorithm
previous_step_size = 1 # 
max_iters = 1000 # maximum number of iterations
iters = 0 #iteration counter
df = lambda x: -6.65 * x #Gradient of our function
```

```python
x_list, y_list = [cur_x], [function(cur_x)]
while previous_step_size > precision and iters < max_iters:
    prev_x = cur_x #Store current x value in prev_x
    cur_x = cur_x - rate * df(prev_x) #Grad descent
    previous_step_size = abs(cur_x - prev_x) #Change in x
    iters = iters+1 #iteration count
    
    x_list.append(cur_x)
    y_list.append(function(cur_x))
print("The local minimum occurs at", cur_x)
```

```python
plt.scatter(x_list,y_list,c="g")
plt.plot(x_list,y_list,c="g")
plt.title("Gradient descent")
plt.show()
```
